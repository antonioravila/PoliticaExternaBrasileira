#Análise de Discurso - Política Externa Brasileira



---


* O atual notebook busca análisar os discursos dos quatro primeiros presidentes da Ditadura Militar Brasileira de 1964 no que se refere à política externa brasileira.
* Dessa maneira, foi compiladas as Menssagens ao Congresso Nacional de cada presidente durante seu mandato. Tais documentos são feitos com frequência anual, e foram enviados durante grande parte da história republicana brasileira, tendo não sido somente durante períodos específicos, como por exemplo a ditadura de Getúlio Vargas. Eles tem como foco: “o objetivo de prestar contas das ações do Executivo do ano anterior e apresentar
planos e expectativas para o ano vigente.” (REPÚBLICA, 2020, s.p).


## Bibliotecas

```python
# Importando bibliotecas necessárias para a análise dos discursos
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

## Dados

```python
# Importando o compilado dos Discursos 
discursos_castelo = open('Castelo Branco.txt', encoding='cp1252').read()
discursos_costa_e_silva = open('Costa e Silva.txt', encoding='cp1252').read()
discursos_medici = open('Medici.txt', encoding='cp1252').read()
discursos_geisel = open('Geisel.txt', encoding='cp1252').read()
```

##Wordclouds

### Stopwords

```python
# Definindo algumas palavras que serão excluidas da wordcloud.
stopwords = set(STOPWORDS)
stopwords.update(["nossa", "dois", "seu", "nos", "está", "num", "têm", "sua", 
                  "tem", "deu", "Ainda", "de", "da", "para", "em", "na", "por", 
                  "das", "uma", "que", "mais", "ao", "se", "dos", "os", "nosso", 
                  "não", "cada", "entre", "aos", "quais", "mesmo", "todo", "um",   
                  "foi", "quanto", "qual", "teve", "mas", "seja", "mas", "ou",
                  "tendo", "nossa","como", "essa", "são", "sobre", "também","às",
                  "pela", "além", "Brasil", "países"])
```

### Castelo Branco

```python
text = discursos_castelo 

wordcloud = WordCloud(stopwords=stopwords, max_font_size=150, max_words=100, 
              background_color="white", width=800, height=400).generate(text)

fig = plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

fig.savefig("castelo_branco.png")
```
